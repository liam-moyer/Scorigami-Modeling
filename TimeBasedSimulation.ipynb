```python
#OT Score

def ot_scoring(t1, t2, avg_mu):
    ot = pd.read_csv("All NFL Game Data (Including OT Stats).csv")

    keylist = ['0-0', '2-0', '3-0', '6-0', '3-3', '6-3']

    ot_count=0

    ot_scores = {}
    for i in keylist:
        ot_scores[i] = 0

    w1=0
    w2=0
    for period in range(6):
        w1+=(avg_mu[t1][period][0]+avg_mu[t1][period][1]+avg_mu[t1][period][2]+avg_mu[t1][period][3]+avg_mu[t1][period][4]+avg_mu[t1][period][5])
        w2+=(avg_mu[t2][period][0]+avg_mu[t2][period][1]+avg_mu[t2][period][2]+avg_mu[t2][period][3]+avg_mu[t2][period][4]+avg_mu[t2][period][5])

    weight1=w1/(w1+w2)
    weight2=1-weight1



    scoring_team = random.choices([1,2], weights=(weight1, weight2))
    scoring_team = scoring_team[0]



    for index, row in ot.iterrows():
        if (row['SEASON']>=2015 and row['OVERTIME']):
            ot_count+=1
            ot_scores[row['OT SCORE']]+=1

    for score in ot_scores:
        ot_scores[score]=ot_scores[score]/2



    ot_event = random.choices([1,2,3,4,5,6], weights=(ot_scores['0-0']/ot_count, ot_scores['2-0']/ot_count, ot_scores['3-0']/ot_count, ot_scores['6-0']/ot_count, ot_scores['3-3']/ot_count, ot_scores['6-3']/ot_count))
    ot_event = ot_event[0]


    if ot_event == 1:
        return (scoring_team, [0,0])
    if ot_event == 2:
        return (scoring_team, [2,0])
    if ot_event == 3:
        return (scoring_team, [3,0])
    if ot_event == 4:
        return (scoring_team, [6,0])
    if ot_event == 5:
        return (scoring_team, [3,3])
    if ot_event == 6:
        return (scoring_team, [6,3])
```

```python
# Modified Time Based With Team Strength

def simulation5(avg_mu):

        stoplist = [15, 28, 30, 45, 58, 60]
        numbers = list(range(0, 32))
        random.shuffle(numbers)

        t1=numbers[0]
        t2=numbers[1]

        #print('t1 = ', t1, ' t2 = ', t2)

        t=0
        s1=0
        s2=0
        x=0

        for time in stoplist:

            scoring_events=0

            #First Quarter
            if (time==15):
                x = np.random.poisson(period1_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=0)
                    scoring_events+=1

            #Second Quarter
            elif (time==28):
                x = np.random.poisson(period2_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=1)
                    scoring_events+=1


            #Second Quarter 2 min remaining
            elif (time==30):
                x = np.random.poisson(period3_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=2)
                    scoring_events+=1

            #Third Quarter
            elif (time==45):
                x = np.random.poisson(period4_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=3)
                    scoring_events+=1


            #Fourth Quarter
            elif (time==58):
                x = np.random.poisson(period5_mu)

                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=4)
                    scoring_events+=1


            #Fourth Quarter 2 minutes remaining
            elif (time==60):

                x = np.random.poisson(period6_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=5)
                    scoring_events+=1


        if (time==60 and s1 != s2):
            #print('Final Score: ', s1, '-', s2)
            return([s1,s2])

        else:
            #print('OT - Current score is: ', s1, '-', s2)
            ot_score_updated = ot_scoring(t1, t2, avg_mu)

            #T1 Scores
            if ot_score_updated[0]==1:
                s1 += ot_score_updated[1][0]
                s2 += ot_score_updated[1][1]
            else:
                s1 += ot_score_updated[1][1]
                s2 += ot_score_updated[1][0]
            #print('OT final score is: ', s1, '-', s2)
            return([s1,s2])
```

```python
# Score Event Team Strength

def score_event_ts(s1, s2, t1, t2, avg_mu, period):

    XP_or_2PT=0

    #print((avg_mu[t1][period][0]+avg_mu[t1][period][1]+avg_mu[t1][period][2]+avg_mu[t1][period][3]+avg_mu[t1][period][4]+avg_mu[t1][period][5]+avg_mu[t2][period][0]+avg_mu[t2][period][1]+avg_mu[t2][period][2]+avg_mu[t2][period][3]+avg_mu[t2][period][4]+avg_mu[t2][period][5]))
    w1=0
    w2=0

    td1_rate=0
    fg1_rate=0
    sfty1_rate=0

    td2_rate=0
    fg2_rate=0
    sfty2_rate=0

    if t1<=7:
        for x in range(8):
            w1+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td1_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg1_rate+=avg_mu[x][period][3]
            sfty1_rate+=avg_mu[x][period][4]

    elif t1<=15:
        for x in range(8, 16):
            w1+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td1_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg1_rate+=avg_mu[x][period][3]
            sfty1_rate+=avg_mu[x][period][4]

    elif t1<=23:
        for x in range(16, 24):
            w1+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td1_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg1_rate+=avg_mu[x][period][3]
            sfty1_rate+=avg_mu[x][period][4]

    elif t1<=31:
        for x in range(24, 32):
            w1+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td1_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg1_rate+=avg_mu[x][period][3]
            sfty1_rate+=avg_mu[x][period][4]

    if t2<=7:
        for x in range(8):
            w2+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td2_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg2_rate+=avg_mu[x][period][3]
            sfty2_rate+=avg_mu[x][period][4]

    elif t2<=15:
        for x in range(8, 16):
            w2+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td2_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg2_rate+=avg_mu[x][period][3]
            sfty2_rate+=avg_mu[x][period][4]

    elif t2<=23:
        for x in range(16, 24):
            w2+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td2_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg2_rate+=avg_mu[x][period][3]
            sfty2_rate+=avg_mu[x][period][4]

    elif t2<=31:
        for x in range(24, 32):
            w2+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td2_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg2_rate+=avg_mu[x][period][3]
            sfty2_rate+=avg_mu[x][period][4]

    w1=w1*1/8
    w2=w2*1/8

    td1_rate=td1_rate*1/8
    fg1_rate=fg1_rate*1/8
    sfty1_rate=sfty1_rate*1/8

    td2_rate=td2_rate*1/8
    fg2_rate=fg2_rate*1/8
    sfty2_rate=sfty2_rate*1/8

    weight1=w1/(w1+w2)
    weight2=1-weight1

    scoring_team = random.choices([1,2], weights=(weight1, weight2))
    scoring_team = scoring_team[0]

    #if scoring_team==1:
        #rank=t1
    #else:
        #rank=t2

    if scoring_team==1:
        event = random.choices(["FG", "TD", "SFTY"], weights=(fg1_rate, td1_rate, sfty1_rate))
    elif scoring_team==2:
        event = random.choices(["FG", "TD", "SFTY"], weights=(fg2_rate, td2_rate, sfty2_rate))

    event=event[0]

   #Field Goal
    if(event=='FG'):
        if (scoring_team==1):
            return[s1+3, s2]
        else:
            return[s1, s2+3]

    #Safety
    elif(event=='SFTY'):
        if (scoring_team==1):
            return[s1+2, s2]
        else:
            return[s1, s2+2]

    #Touchdown
    else:

        if(scoring_team==1):

            #If score difference does not exist kick xp
            if((s1+6)-s2 not in xp_attempt_perc.keys()):
                XP_or_2PT=1

            else:
                #Deciding 1 or 2 point
                XP_or_2PT_Choice = random.choices([1,2], weights=(xp_attempt_perc[(s1+6)-s2], xp2_attempt_perc[(s1+6)-s2]))
                XP_or_2PT=XP_or_2PT_Choice[0]

            #If 1 point, calculate make rate
            if (XP_or_2PT == 1):
                make_xp = random.choices([1,0], weights=(xp_make_perc, 1-xp_make_perc))
                make_xp=make_xp[0]
                return [s1+6+make_xp, s2]


            #If 2 points, calculate make rate
            elif (XP_or_2PT == 2):
                make_xp2 = random.choices([2,0], weights=(xp2_make_perc, 1-xp2_make_perc))
                make_xp2=make_xp2[0]
                return [s1+6+make_xp2, s2]

        #Team 2 Scores
        else:

            #If score difference does not exist kick xp
            if((s2+6)-s1 not in xp_attempt_perc.keys()):
                XP_or_2PT=1

            else:
                #Deciding 1 or 2 point
                XP_or_2PT_Choice = random.choices([1,2], weights=(xp_attempt_perc[(s2+6)-s1], xp2_attempt_perc[(s2+6)-s1]))
                XP_or_2PT=XP_or_2PT_Choice[0]

            #If 1 point, calculate make rate
            if (XP_or_2PT == 1):
                make_xp = random.choices([1,0], weights=(xp_make_perc, 1-xp_make_perc))
                make_xp=make_xp[0]
                return [s1, s2+6+make_xp]


            #If 2 points, calculate make rate
            elif (XP_or_2PT == 2):
                make_xp2 = random.choices([2,0], weights=(xp2_make_perc, 1-xp2_make_perc))
                make_xp2=make_xp2[0]
                return [s1, s2+6+make_xp2]

    return [s1, s2]
```

```python
# Modified Time Based With Team Strength

def simulation6(avg_mu):

        stoplist = [15, 28, 30, 45, 58, 60]
        numbers = list(range(0, 32))
        random.shuffle(numbers)

        t1=numbers[0]
        t2=numbers[1]

        t=0
        s1=0
        s2=0
        x=0

        #Get Mu's
        t1_scoring_avg=0
        t2_scoring_avg=0
        combined_avg=0
        for period in range(6):
            t1_scoring_avg += avg_mu[t1][period][0]+avg_mu[t1][period][1]+avg_mu[t1][period][2]+avg_mu[t1][period][3]+avg_mu[t1][period][4]+avg_mu[t1][period][5]
            t2_scoring_avg += avg_mu[t2][period][0]+avg_mu[t2][period][1]+avg_mu[t2][period][2]+avg_mu[t2][period][3]+avg_mu[t2][period][4]+avg_mu[t2][period][5]

        combined_avg=(t1_scoring_avg+t2_scoring_avg)/2

        #mu 1
        mu1 = combined_avg*(q1_rate)*4

        # mu2
        mu2 = combined_avg*(q2_rate)*(60/13)

        # mu3
        mu3 = combined_avg*(q2_2min_rate)*(60/2)

        # mu4
        mu4 = combined_avg*(q3_rate)*4

        # mu5
        mu5 = combined_avg*(q4_rate)*(60/13)

        # mu6
        mu6 = combined_avg*(q4_2min_rate)*(60/2)

        for time in stoplist:

            scoring_events=0

            #First Quarter
            if (time==15):

                x = np.random.exponential(1/mu1)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=0)
                    scoring_events+=1

            #Second Quarter
            elif (time==28):
                x = np.random.exponential(1/mu2)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=1)
                    scoring_events+=1


            #Second Quarter 2 min remaining
            elif (time==30):
                x = np.random.exponential(1/mu3)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=2)
                    scoring_events+=1

            #Third Quarter
            elif (time==45):
                x = np.random.exponential(1/mu4)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=3)
                    scoring_events+=1


            #Fourth Quarter
            elif (time==58):
                x = np.random.exponential(1/mu5)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=4)
                    scoring_events+=1


            #Fourth Quarter 2 minutes remaining
            elif (time==60):
                x = np.random.exponential(1/mu6)
                while scoring_events < x:
                    [s1, s2] = score_event_ts(s1, s2, t1, t2, avg_mu, period=5)
                    scoring_events+=1


        if (time==60 and s1 != s2):
            return([s1,s2])

        else:
            ot_score_updated = ot_scoring(t1, t2, avg_mu)

            #T1 Scores
            if ot_score_updated[0]==1:
                s1 += ot_score_updated[1][0]
                s2 += ot_score_updated[1][1]
            else:
                s1 += ot_score_updated[1][1]
                s2 += ot_score_updated[1][0]
            #print('OT final score is: ', s1, '-', s2)
            return([s1,s2])
```

```python
# Modified Time Based With Team Strength

def flatsim(avg_mu):

        stoplist = [15, 28, 30, 45, 58, 60]
        numbers = list(range(0, 32))
        random.shuffle(numbers)

        t1=numbers[0]
        t2=numbers[1]

        #print('t1 = ', t1, ' t2 = ', t2)

        t=0
        s1=0
        s2=0
        x=0

        for time in stoplist:

            scoring_events=0

            #First Quarter
            if (time==15):
                x = np.random.poisson(period1_mu)
                while scoring_events < x:
                    [s1, s2] = flat_score_event_ts(s1, s2, t1, t2, avg_mu, period=0)
                    scoring_events+=1

            #Second Quarter
            elif (time==28):
                x = np.random.poisson(period2_mu)
                while scoring_events < x:
                    [s1, s2] = flat_score_event_ts(s1, s2, t1, t2, avg_mu, period=1)
                    scoring_events+=1


            #Second Quarter 2 min remaining
            elif (time==30):
                x = np.random.poisson(period3_mu)
                while scoring_events < x:
                    [s1, s2] = flat_score_event_ts(s1, s2, t1, t2, avg_mu, period=2)
                    scoring_events+=1

            #Third Quarter
            elif (time==45):
                x = np.random.poisson(period4_mu)
                while scoring_events < x:
                    [s1, s2] = flat_score_event_ts(s1, s2, t1, t2, avg_mu, period=3)
                    scoring_events+=1


            #Fourth Quarter
            elif (time==58):
                x = np.random.poisson(period5_mu)

                while scoring_events < x:
                    [s1, s2] = flat_score_event_ts(s1, s2, t1, t2, avg_mu, period=4)
                    scoring_events+=1


            #Fourth Quarter 2 minutes remaining
            elif (time==60):

                x = np.random.poisson(period6_mu)
                while scoring_events < x:
                    [s1, s2] = flat_score_event_ts(s1, s2, t1, t2, avg_mu, period=5)
                    scoring_events+=1


        if (time==60 and s1 != s2):
            #print('Final Score: ', s1, '-', s2)
            return([s1,s2])

        else:
            #print('OT - Current score is: ', s1, '-', s2)
            ot_score_updated = ot_scoring(t1, t2, avg_mu)

            #T1 Scores
            if ot_score_updated[0]==1:
                s1 += ot_score_updated[1][0]
                s2 += ot_score_updated[1][1]
            else:
                s1 += ot_score_updated[1][1]
                s2 += ot_score_updated[1][0]
            #print('OT final score is: ', s1, '-', s2)
            return([s1,s2])
```

```python
# Score Event Team Strength

def flat_score_event_ts(s1, s2, t1, t2, avg_mu, period):

    XP_or_2PT=0

    #print((avg_mu[t1][period][0]+avg_mu[t1][period][1]+avg_mu[t1][period][2]+avg_mu[t1][period][3]+avg_mu[t1][period][4]+avg_mu[t1][period][5]+avg_mu[t2][period][0]+avg_mu[t2][period][1]+avg_mu[t2][period][2]+avg_mu[t2][period][3]+avg_mu[t2][period][4]+avg_mu[t2][period][5]))
    w1=0
    w2=0

    td1_rate=0
    fg1_rate=0
    sfty1_rate=0

    td2_rate=0
    fg2_rate=0
    sfty2_rate=0

    for x in range(32):
        for period in range(6):
            w1+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td1_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg1_rate+=avg_mu[x][period][3]
            sfty1_rate+=avg_mu[x][period][4]

            w2+=(avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][3]+avg_mu[x][period][4]+avg_mu[x][period][5])
            td2_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
            fg2_rate+=avg_mu[x][period][3]
            sfty2_rate+=avg_mu[x][period][4]

    w1=(w1*1/32)/6
    w2=(w2*1/32)/6

    td1_rate=(td1_rate*1/32)/6
    fg1_rate=(fg1_rate*1/32)/6
    sfty1_rate=(sfty1_rate*1/32)/6

    td2_rate=(td2_rate*1/32)/6
    fg2_rate=(fg2_rate*1/32)/6
    sfty2_rate=(sfty2_rate*1/32)/6

    weight1=w1/(w1+w2)
    weight2=1-weight1

    scoring_team = random.choices([1,2], weights=(weight1, weight2))
    scoring_team = scoring_team[0]

    #if scoring_team==1:
        #rank=t1
    #else:
        #rank=t2

    if scoring_team==1:
        event = random.choices(["FG", "TD", "SFTY"], weights=(fg1_rate, td1_rate, sfty1_rate))
    elif scoring_team==2:
        event = random.choices(["FG", "TD", "SFTY"], weights=(fg2_rate, td2_rate, sfty2_rate))

    event=event[0]

   #Field Goal
    if(event=='FG'):
        if (scoring_team==1):
            return[s1+3, s2]
        else:
            return[s1, s2+3]

    #Safety
    elif(event=='SFTY'):
        if (scoring_team==1):
            return[s1+2, s2]
        else:
            return[s1, s2+2]

    #Touchdown
    else:

        if(scoring_team==1):

            #If score difference does not exist kick xp
            if((s1+6)-s2 not in xp_attempt_perc.keys()):
                XP_or_2PT=1

            else:
                #Deciding 1 or 2 point
                XP_or_2PT_Choice = random.choices([1,2], weights=(xp_attempt_perc[(s1+6)-s2], xp2_attempt_perc[(s1+6)-s2]))
                XP_or_2PT=XP_or_2PT_Choice[0]

            #If 1 point, calculate make rate
            if (XP_or_2PT == 1):
                make_xp = random.choices([1,0], weights=(xp_make_perc, 1-xp_make_perc))
                make_xp=make_xp[0]
                return [s1+6+make_xp, s2]


            #If 2 points, calculate make rate
            elif (XP_or_2PT == 2):
                make_xp2 = random.choices([2,0], weights=(xp2_make_perc, 1-xp2_make_perc))
                make_xp2=make_xp2[0]
                return [s1+6+make_xp2, s2]

        #Team 2 Scores
        else:

            #If score difference does not exist kick xp
            if((s2+6)-s1 not in xp_attempt_perc.keys()):
                XP_or_2PT=1

            else:
                #Deciding 1 or 2 point
                XP_or_2PT_Choice = random.choices([1,2], weights=(xp_attempt_perc[(s2+6)-s1], xp2_attempt_perc[(s2+6)-s1]))
                XP_or_2PT=XP_or_2PT_Choice[0]

            #If 1 point, calculate make rate
            if (XP_or_2PT == 1):
                make_xp = random.choices([1,0], weights=(xp_make_perc, 1-xp_make_perc))
                make_xp=make_xp[0]
                return [s1, s2+6+make_xp]


            #If 2 points, calculate make rate
            elif (XP_or_2PT == 2):
                make_xp2 = random.choices([2,0], weights=(xp2_make_perc, 1-xp2_make_perc))
                make_xp2=make_xp2[0]
                return [s1, s2+6+make_xp2]

    return [s1, s2]
```

```python
# Scoring Rates Simulation

def scorediff_sim(avg_mu):

        stoplist = [15, 28, 30, 45, 58, 60]
        numbers = list(range(0, 32))
        random.shuffle(numbers)

        t1=numbers[0]
        t2=numbers[1]

        t=0
        s1=0
        s2=0
        x=0

        for time in stoplist:

            scoring_events=0

            #First Quarter
            if (time==15):
                x = np.random.poisson(period1_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_sd(s1, s2, t1, t2, avg_mu, period=0)
                    scoring_events+=1

            #Second Quarter
            elif (time==28):
                x = np.random.poisson(period2_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_sd(s1, s2, t1, t2, avg_mu, period=1)
                    scoring_events+=1


            #Second Quarter 2 min remaining
            elif (time==30):
                x = np.random.poisson(period3_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_sd(s1, s2, t1, t2, avg_mu, period=2)
                    scoring_events+=1

            #Third Quarter
            elif (time==45):
                x = np.random.poisson(period4_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_sd(s1, s2, t1, t2, avg_mu, period=3)
                    scoring_events+=1


            #Fourth Quarter
            elif (time==58):
                x = np.random.poisson(period5_mu)

                while scoring_events < x:
                    [s1, s2] = score_event_sd(s1, s2, t1, t2, avg_mu, period=4)
                    scoring_events+=1


            #Fourth Quarter 2 minutes remaining
            elif (time==60):

                x = np.random.poisson(period6_mu)
                while scoring_events < x:
                    [s1, s2] = score_event_sd(s1, s2, t1, t2, avg_mu, period=5)
                    scoring_events+=1


        if (time==60 and s1 != s2):
            return([s1,s2])

        else:
            ot_score_updated = ot_scoring(t1, t2, avg_mu)

            #T1 Scores
            if ot_score_updated[0]==1:
                s1 += ot_score_updated[1][0]
                s2 += ot_score_updated[1][1]
            else:
                s1 += ot_score_updated[1][1]
                s2 += ot_score_updated[1][0]
            return([s1,s2])
```

```python
def score_event_sd(s1, s2, t1, t2, avg_mu, period):

    w1=0
    w2=0

    td_rate=0
    fg_rate=0
    sfty_rate=0

    #Time Remaining at start of each period
    time_remaining_list = [60, 45, 32, 30, 13, 2]
    time_remaining = time_remaining_list[period]

    #Hard coded values of delta and lambda
    delta=0.1
    lamb=1

    theta = pow(2.718281828459045, ((lamb*-1)*(60-time_remaining)))

    # Flat over team rank
    for x in range(32):
        td_rate+=avg_mu[x][period][0]+avg_mu[x][period][1]+avg_mu[x][period][2]+avg_mu[x][period][5]
        fg_rate+=avg_mu[x][period][3]
        sfty_rate+=avg_mu[x][period][4]

    td_rate=td_rate/32
    fg_rate=fg_rate/32
    sfty_rate=sfty_rate/32

    #Full Flat
    mu_bar = 0
    for team in range(32):
        for event in range(6):
            mu_bar+=(avg_mu[team][period][event])
    mu_bar = (mu_bar*1/32)

    #Mu 1 and Mu 2
    mu_1=0
    for event in range(6):
        mu_1+=avg_mu[t1][period][event]

    mu_2=0
    for event in range(6):
        mu_2+=avg_mu[t2][period][event]


    w1=0
    w2=0

    if s1>s2:
        w1 = mu1*theta+(1-theta)*mu_bar*(1-delta)
        w2 = mu2*theta+(1-theta)*mu_bar*(1+delta)
    elif s1<s2:
        w1 = mu1*theta+(1-theta)*mu_bar*(1+delta)
        w2 = mu2*theta+(1-theta)*mu_bar*(1-delta)
    else:
        w1 = mu1*theta+(1-theta)*mu_bar
        w2 = mu2*theta+(1-theta)*mu_bar

    weight1=(w1/(w1+w2))
    weight2=1-weight1

    #Flip Coin
    scoring_team = random.choices([1,2], weights=(weight1, weight2))
    scoring_team = scoring_team[0]


    XP_or_2PT=0

    event = random.choices(["FG", "TD", "SFTY"], weights=(fg_rate, td_rate, sfty_rate))
    event=event[0]

   #Field Goal
    if(event=='FG'):
        if (scoring_team==1):
            return[s1+3, s2]
        else:
            return[s1, s2+3]

    #Safety
    elif(event=='SFTY'):
        if (scoring_team==1):
            return[s1+2, s2]
        else:
            return[s1, s2+2]

    #Touchdown
    else:

        if(scoring_team==1):

            #If score difference does not exist kick xp
            if((s1+6)-s2 not in xp_attempt_perc.keys()):
                XP_or_2PT=1

            else:
                #Deciding 1 or 2 point
                XP_or_2PT_Choice = random.choices([1,2], weights=(xp_attempt_perc[(s1+6)-s2], xp2_attempt_perc[(s1+6)-s2]))
                XP_or_2PT=XP_or_2PT_Choice[0]

            #If 1 point, calculate make rate
            if (XP_or_2PT == 1):
                make_xp = random.choices([1,0], weights=(xp_make_perc, 1-xp_make_perc))
                make_xp=make_xp[0]
                return [s1+6+make_xp, s2]


            #If 2 points, calculate make rate
            elif (XP_or_2PT == 2):
                make_xp2 = random.choices([2,0], weights=(xp2_make_perc, 1-xp2_make_perc))
                make_xp2=make_xp2[0]
                return [s1+6+make_xp2, s2]

        #Team 2 Scores
        else:

            #If score difference does not exist kick xp
            if((s2+6)-s1 not in xp_attempt_perc.keys()):
                XP_or_2PT=1

            else:
                #Deciding 1 or 2 point
                XP_or_2PT_Choice = random.choices([1,2], weights=(xp_attempt_perc[(s2+6)-s1], xp2_attempt_perc[(s2+6)-s1]))
                XP_or_2PT=XP_or_2PT_Choice[0]

            #If 1 point, calculate make rate
            if (XP_or_2PT == 1):
                make_xp = random.choices([1,0], weights=(xp_make_perc, 1-xp_make_perc))
                make_xp=make_xp[0]
                return [s1, s2+6+make_xp]


            #If 2 points, calculate make rate
            elif (XP_or_2PT == 2):
                make_xp2 = random.choices([2,0], weights=(xp2_make_perc, 1-xp2_make_perc))
                make_xp2=make_xp2[0]
                return [s1, s2+6+make_xp2]

    return [s1, s2]
```
